## Masking(removing background of an image and pasting the image into another image)

```python
from PIL import Image

image_copy = Image.open("/content/drive/MyDrive/computer vision/curry.jpg").convert("RGBA")
logo = Image.open("/content/drive/MyDrive/computer vision/car.jpg").convert("RGBA")  # Ensure logo has an alpha channel

# Compute position for bottom-right corner
position = (image_copy.width - logo.width, image_copy.height - logo.height)
image_copy.paste(logo, position, logo.split()[3])  # Use the alpha channel as mask

from IPython.display import display
display(image_copy) # The masked image will be displayed at the bottom corner as the coordinates are mentioned such
```

## Drawing an image using PIL

```python
from PIL import  ImageDraw
```

```python
image = Image.new('RGB', (400, 400)) # assigning image variable and it's properties with RGB format size 400x400
img_draw = ImageDraw.Draw(image) #constructs the image with above properties
```

```python
img_draw.rectangle((100, 30, 300, 200), outline='red', fill='white') # cropped frame , border of frame, color inside frame
img_draw.text((150, 100), 'Om Wadhwani', fill='red') # text alignment, text , color of text
image.save('drawing.jpg')
```
